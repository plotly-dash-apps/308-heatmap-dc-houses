```python
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

import plotly as py
import plotly.graph_objs as go
from plotly.graph_objs import *
```

```python
df = pd.read_csv('Crime_Incidents_in_2019.csv')
df.head(3)
```

```python
df2=df.groupby(['LATITUDE', 'LONGITUDE'])['REPORT_DAT'].count().reset_index()
df2.columns=['LATITUDE', 'LONGITUDE', 'COUNT']
df2.head(3)
```

```python
df2.to_csv('crime_shortdf.csv', index=False)
```

```python
crimes =pd.read_csv('crime_shortdf.csv')
crimes.head()
```

```python
# Display as a heatmap 

fig = go.Figure(go.Densitymapbox(lat=crimes['LATITUDE'], 
                                 lon=crimes['LONGITUDE'], 
                                 z=crimes['COUNT'], 
                                 radius=20))
fig.update_layout(mapbox_style="stamen-terrain", 
                  mapbox_center_lon=-77.07,
                  mapbox_center_lat=38.92,
                  mapbox_zoom=11,
                 )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
iplot(fig)
```
